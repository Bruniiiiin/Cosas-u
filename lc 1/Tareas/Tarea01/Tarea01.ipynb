#### b) Reporte el resultado de la medición del tiempo que tarda en caer la bola con su error correspondiente:

Como buscamos estimar un valor real de cuanto tiempo demora en caer la pelota utilizaremos el error del promedio. 

Primero calculamos el promedio del tiempo:

$\begin{align}
    \overline{t} &= \frac{1}{N} \sum_{i=1}^{N} t_{i} \\  
                 &= \frac{1}{10} \sum_{i=1}^{10} t_{i} \\
    \overline{t} &= 0.659
 \end{align}
$

Teniendo en cuenta que los valores del tiempo medido se nos entregan con dos cifras significativas, decimos que $ \overline{t} \approx 0.66 $ ahora nosotros queremos encontrar el error asociado a este valor final del tiempo que se demora la pelota en caer, por lo que utilizaremos el error estandar del promedio ya que queremos acercarnos al valor real del tiempo dependiendo del numero de mediciones, de todas formas debemos calcular la desviacion estandar, que será:

$
\begin{align}
    \sigma &= \sqrt{\frac{\sum_{i=1}^{N}(t_i - \overline{t})^2}{N}} \\
           &= \sqrt{\frac{\sum_{i=1}^{10}(t_i - 0.66)}{10}} \\
    \sigma &= 0.02449...
\end{align}
$

Por lo que tomando solo dos cifras significativas tenemos que $ \sigma \approx 0.02 $ y utilizando este valor podremos calcular el error estandar del promedio:

$
\begin{align}
    \delta t &= \frac{\sigma}{\sqrt{N}} \\
             &= \frac{0.02}{\sqrt{10}} \\
    \delta t &= 0.00632... 
\end{align}
$

Podemos notar que el valor de nuestro error tiene mas cifras significativas por lo que diremos que el valor del error será de $ \delta t \approx 0.01 $

Por lo que la pelota cae en el siguiente tiempo dado:

$
\begin{equation}
    t = 0.66 \pm 0.01 [s]
\end{equation}
$

#### c) Reporte el resultado de la medición de la aceleracion de gravedad con su error correspondiente:

Para calcular la aceleración de gravedad utilizaremos el siguiente modelo teorico:

$
\begin{align}
    h = \frac{1}{2}g t^2 \\
    g = \frac{2h}{t^2} \\
\end{align}
$

* $h$ : altura
* $g$ : aceleración de gravedad
* $t$ : tiempo

Gracias a esto podemos definir $g$ como función de $h$ y $t$, de esta forma podemos calcular el error propagado a $g(h,t)$ de forma más simple gracias a esta definición:

$
\begin{align}
    g(h, t) = \frac{2h}{t^2}
\end{align}
$

luego:

$
\frac{\partial g}{\partial h} = \frac{2}{t^2} \qquad
\frac{\partial g}{\partial t} = \frac{-4h}{t^3}
$

Esto para utilizar lo siguiente:

$
\begin{align}
    (\Delta f)^2 &= \left(\frac{\partial f}{\partial x} \Delta x \right)^2 + \left(\frac{\partial f}{\partial y} \Delta y \right)^2 \dots  \\
\end{align}
$

Entonces:

$
\begin{align}
    (\Delta g)^2 &= \left(\frac{\partial g}{\partial h} \Delta h\right)^2 + \left(\frac{\partial g}{\partial t} \Delta t\right)^2 \\
                 &= \left(\frac{2}{t^2} \Delta h \right)^2 + \left(\frac{-4h}{t^3} \Delta t \right)^2 \\
                 &= \left(\frac{-4h}{t^3} \Delta t \right)^2 \\
                 &= \left| \frac{-4h}{t^3} \Delta t \right| \\
                 &= \left| \frac{-8}{(0.66)^3} \cdot 0.01  \right| \\
                 &= \left| -27.82 \cdot 0.01 \right| \\
                 &= \left| -0.2782 \right| \\
        \Delta g &= 0.2782
\end{align}
$

Estos calculos son correctos si tenemos en cuenta que la cinta mide en efecto $2[m]$ y que no hay un error en este. De la misma forma que con el tiempo, tomamos solo dos cifras significativas por lo que diremos que el valor del error propagado en $g$ será de $ \Delta g \approx 0.27 $

Ahora calcularemos el promedio de $g$ utilizando el tiempo promedio.

$
\begin{align}
    \overline g &= \frac{2h}{\overline t^2} \\
                &= \frac{4}{0.4356} \\
    \overline g &= 9.182736 \dots
\end{align}
$

Redondeando a nuestra cantidad de cifras significativas tenemos que $ \overline g \approx 9.18 $ . LLegando al resutado de que nuestro valor aproximado de la aceleración de gravedad con su error correspondiente estará dado por:

$
\begin{equation}
    g = 9.18 \pm 0.27  \frac{[m]}{[s]^2}
\end{equation}
$

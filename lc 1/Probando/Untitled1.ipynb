```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = int(input("n ="))
num = 10*np.random.rand(n)
print(num)
c = np.histogram(num, bins=range(0,11))[0]
plt.bar(np.arange(0.5,10,1), c, width=1)
plt.xlabel("Numeros aleatorios")
plt.ylabel("Repeticiones")
```

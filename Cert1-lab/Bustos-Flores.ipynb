```python
import numpy as np
import matplotlib.pyplot as plt


datos = np.genfromtxt("Temperatura-vs-tiempo.txt")

t = datos[:,0]
tem = datos[:,1]

err_t = 0.01
err_tem = 1

plt.errorbar(t, tem, xerr=err_t, yerr=err_tem, fmt="o", label="Datos experimentales")
plt.xlabel("Tiempo")
plt.ylabel("Temperatura")
plt.title("Temperatura vs Tiempo")
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


datos = np.genfromtxt("Temperatura-vs-tiempo.txt")

t = datos[:,0]
tem = datos[:,1]
tem_a = 98
valor = np.log(tem)/tem_a

err_t = 0.01
err_tem = 1

coef_polinomios = np.polyfit(t, valor, 1)

polinomio = np.poly1d(coef_polinomios)

u = np.linspace(min(t), max(t), 100)

v = polinomio(u)

plt.plot(u, v)
plt.errorbar(t, valor, xerr=err_t, yerr=err_tem, fmt="o", label="Datos experimentales")
plt.xlabel("Tiempo")
plt.ylabel("Temperatura")
plt.title("Temperatura vs Tiempo")
plt.grid()
plt.show()
```
